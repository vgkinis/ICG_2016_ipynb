### Atoms nuclei consist of two kinds of particles; protons and neutrons - they roughly have the same mass. Protons are positively charged.
* Atomic Number Z: number of protons
* N: The number of neutrons
* The mass number A is the sum of protons and neutrons (nucleons) and $A = Z+A$

### Elements with the same $Z$ but different $A$ are called **isotopes**.

### The Atomic number $Z$ typically determines the chemical properties of an element. 

* For the *light elements* the most abundant nuclides lie on the $Z = N$ line and are stable $\left(^{2}_{1}\text{H}_2\,,\, ^{4}_{2}\text{He}_2\,,\, ^{12}_{\,\,6}\text{C}_6 \,,\, ^{14}_{\,\,7}\text{N}_7 \,,\, ^{16}_{\,\,8}\text{O}_8\right)$. 

* A slight deviation from the $Z = N$ line is also possible for stable nuclei $\left(^{17}_{\,\,8}\text{O}_9 \,,\,^{18}_{\,\,8}\text{O}_{10}\right)$. 

* Assymetric nuclei typically have lower natural abundances.

* Oxygen Triplet: $^{16}\text{O}\,,\,^{17}\text{O}\,,\,^{18}\text{O}:\,0.9976\,,\,0.00038\,,\,0.00205$
* Hydrogen : $^{1}\text{H}\,,\,^{2}\text{H}:\,0.99985, \,0.00015$

# Abundance and Fractionation

## Isotopic Abundance 

Isotopic Abundances are defined as:
$R = \frac{\text{abundance of rare isotope}}{\text{abundance of abundant isotope}}$

so we can define
$^{18}\text{R(H}_2\text{O}) = \frac{\left[\text{H}_2{}^{18}\text{O}\right]}{\left[\text{H}_2{}^{16}\text{O}\right]}$, 
$^{2}\text{R(H}_2\text{O}) = \frac{\left[\text{HD}{}^{16}\text{O}\right]}{2\left[\text{H}_2{}^{16}\text{O}\right]}$

This is **not** an isotope concentration (molar ratio). This would be defined as:
$\frac{\left[\text{H}_2{}^{18}\text{O}\right]}{\left[\text{H}_2{}^{18}\text{O}\right]+ \left[\text{H}_2{}^{16}\text{O}\right]}$

Another (not so) common expression that can be find in the literature is the atom%. This is defined as:
$\frac{N_X}{N_{tot}}\cdot100\%$, where $N_X$ the number of atoms of the isotope X and $N_{tot}$ the total number of atoms.

# Fractionation

The various *[isotopologues](https://goldbook.iupac.org/I03351.html)* of water present very small differences with respect to their chemical and mostly physical properties. This is a phenomenon called *isotopic fractionation*. Typical examples of processes  where isotopic fractionation can be observed are the transitions between states (liquid to vapor, solid to liquid etc.).

## Reasons for isotopic fractionation
* **Kinetics**
Temperature determines the kinetic energy of a molecule as $kT \sim \frac{1}{2}mv^2$. As a result, at the same $T$ the heavier isotopologues will present lower velocities on average. As a result diffusion velocities and collision frequencies will be lower for the heavier isotopologues.

* **Binding Energies**
Heavier isotopologues generally (not always!) present higher binding energies thus "sitting" deeper in the energy well as seen in the picture below. 

### **Practically this results in:**
* Light isotopologues react faster in chemical reactions.
* Light isotopologues tend to have higher vapor pressures and thus evaporate easier.
* At high temperatures fractionation effects diminish.

# Fractionation Factors - Equilibrium, Diffusion, Kinetic
## Equilibrium Fractionation
Lets assume the following reaction where ${}^{0}\text{X}$ and ${}^{1}\text{X}$ are two different isotopes of the element X.

$\text{A}{}^{0}\text{X} + \text{B}{}^{1}\text{X} \rightleftharpoons \text{A}{}^{1}\text{X} + \text{B}{}^{0}\text{X}$

then the equilibrium constant is $K_X(T) = \frac{\left[\text{A}{}^{1}\text{X}\right]\left[ \text{B}{}^{0}\text{X} \right]  }{\left[ \text{A}{}^{0}\text{X} \right]  \left[ \text{B}{}^{1} \text{X} \right]} = \frac{R_{AX}}{R_{BX}} = \alpha_{AX/BX}$

Here $\alpha_{AX/BX} $ is called the **fractionation factor**.

A very relevant reaction for ice core studies is the under equilibrium evaporation/condensation process of the water molecule:

$ \text{H}_2{}^{16}\text{O}_{vap} +  \text{H}_2{}^{18}\text{O}_{liq}  \leftrightharpoons  \text{H}_2{}^{16}\text{O}_{liq} +  \text{H}_2{}^{18}\text{O}_{vap}   $

For which the fractionation factor of the ${}^{18}\text{O}$ will be:
$K(T) = \frac{\left[\text{H}_2{}^{18}\text{O}_{vap}\right]\left[ \text{H}_2{}^{16}\text{O}_{liq} \right]  }{\left[ \text{H}_2{}^{16}\text{O}_{vap} \right]  \left[ \text{H}_2{}^{18} \text{O}_{liq} \right]} = \frac{{}^{18}R_{vap}}{{}^{18}R_{liq}} = {}^{18}\alpha_{v/l}$

${}^{18}\alpha_{v/l}$ has a value vey close to 1 (remember that especially at higher temperatures the fractionation effects are of very small magnitude). For convenience we also define the *fractionation* ${}^{18}\varepsilon_{v/l} = {}^{18}\alpha_{v/l} - 1$ that has a value very close to zero and it is typically expressed in per mil $\text{\permil}$

It becomes clear that negative values of $\varepsilon$ designate *depletion* in the rare isotope and positive values of $\varepsilon$ designate *enrichment*

# The $\delta$ notation

Generally isotopic abundances are reported using the $\delta$ notation.
$\delta^{18}\text{O} = \left(\frac{{}^{18}\text{R}_{sample}}{{}^{18}\text{R}_{std}} - 1\right) \times 1000 \text{ permile}$

Std here refers to a standard water. [VSMOW](https://en.wikipedia.org/wiki/Vienna_Standard_Mean_Ocean_Water) is considered as the international reference and all isotopic compositions are reported with respect to it.

# The Rayleigh Fractionation Model

```python
Image("../figs/rayleigh_scheme.png", width = 600)
```

### Initial Vapor
* $R_{vo} = \frac{R_0}{\alpha_0}$

### Precipitation Event
* Initial vapor: $N_{vo}$, $R_{vo}$
* Remaining vapor: $N-dN$, $R-dR$
* Formed compound: $dN$, $\alpha R$

### Assuming Mass Balance:
$RN = \left(R - dR\right)\left(N-dN\right) + \alpha R dN$ which approximates to $NdR = \left(\alpha - 1\right)RdN$
we end up with:

$\frac{dR}{R} = \left(\alpha - 1\right) \frac{dN}{N}$

assuming $\alpha$ is independend of $N$ and with initial conditions $R_{vo}$ and $N_0$ we integrate and end up with:

$\frac{R_v}{R_{vo}} = \left( \frac{N}{N_{vo}}  \right)^{\alpha-1}$
